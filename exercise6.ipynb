## 利用数据透视表统计分析销售数据
#### 以下是某有限责任公司一组销售数据，请利用pandas数据透视表相关知识按要求进行统计分析
![Jupyter](https://cdn.py2china.cn/study-group/pandas/6-1.png)
#### 1.将上方数据表建立 DataFrame 进行存储

```python
import pandas as pd
import numpy as np
data = {'id':[1001,1001,1002,1003,1001,1002,1002,1003,1001,1002,1001,1002],
       'date':['2019-01-01','2019-01-01','2019-01-01','2019-01-01','2019-01-02','2019-01-02',
               '2019-01-02','2019-01-02','2019-01-03','2019-01-03','2019-01-04','2019-01-04',],
       'goods':['A','B','A','B','A','A','B','A','A','B','A','A'],
       'sales':[3000,2000,3000,2000,3000,3000,2000,3000,3000,2000,3000,3000]}
df = pd.DataFrame(data)
df
```

#### 2.利用 pivot_table 统计该企业每个人每天销售商品 A 和 B 的金额之和

```python
pd.pivot_table(df,index=['id'],columns = ['date'],aggfunc =np.sum,values='sales')
```

#### 3.请将上面的统计数据再增加一列，统计出每个人的销售总金额

```python
pd.pivot_table(df, index=['id'], columns = ['date'], aggfunc =np.sum, values='sales', margins = True)
```

#### 以下为部分抽烟人群调查表，请将数据存储入 DataFrame，并完成以下操作
![Jupyter](https://cdn.py2china.cn/study-group/pandas/6-2.png)
#### 4.请将以上数据存入 DataFrame 中

```python
data1 = {'sex':['man','man','woman','woman','man','woman','man','woman','woman'],
        'age':[15,23,25,17,35,57,24,31,22],
        'smoke':['True','False','False','True','True','False','False','True','False'],
        'height':[168,179,181,166,173,178,188,190,160]}
df1 = pd.DataFrame(data1)
df1
```

#### 5.请以性别为分组依据，查看男女抽烟平均年龄和身高

```python
table = pd.pivot_table(df1, index=['sex','smoke'])
table.query('smoke == ["True"]')
```

#### 6.利用交叉表知识，统计各个性别抽烟的人数

```python
pd.crosstab(df1.sex,df1.smoke)
```

#### 7.利用交叉表知识，统计各个年龄段抽烟人情况

```python
pd.crosstab(df1.age,df1.smoke)
```
