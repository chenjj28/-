## 上海二手房市场数据分析
#### 以下为上海地区某时期内的二手房数据，存在 houses.csv中，每条房源包含下列信息，请按照要求进行统计分析：
#### 1. 用 pandas 读取 csv 格式文件，并将所有字段的重复数据去除

```python
import pandas as pd
import numpy as np
df = pd.read_csv('houses.csv',encoding = 'gb18030')
df.drop_duplicates().head()
```

#### 2. 请统计出总房源数量、总小区数量、总版块数量（任务提示：直接统计数据的条数，即用 len 函数即可。）

```python
print('总房源数量：',len(set(df['title'])))                   #也可用：len(df['title'].unique())
print('总小区数量：',len(set(df['xiaoqu_name'])))
print('总板块数量：',len(set(df['sub_district_name'])))
```

#### 3. 请统计出有哪些不同的户型、有哪些不同的装修情况、有哪些不同的建造年份（任务提示：可用 unique 函数进行统计。阅读参考文章④）

```python
print('户型种类：', df.huxing.unique())                        #或者写成：df['huxing'].unique()
print('装修类型：', df.zhuangxiu.unique())
print('建造年份：', df.buildyear.unique())
```

#### 4. 请计算出上海二手房每平方米均价

```python
df['up_price'].mean()
```

#### 5. 请统计出不同户型对应的房源数量、平均面积；并从大到小进行排序，判断出主流户型是什么？平均面积多大？（任务提示：以 huxing 为分组对象，展示 size 列，统计户型数量和 size 平均数。）

```python
df.groupby('huxing')['size'].agg(amount = 'count', ave_size = 'mean').sort_values(by='amount', ascending = False)

# Note:
# df.groupby('huxing')['size'].agg({'num_huxing':'count','ave_size':'mean'})这种以dic形式rename列名的写法已经不可再用

# reason: in new pandas version named aggregation is the recommended replacement for the deprecated “dict-of-dicts” 
# approach to naming the output of column-specific aggregations
# (Deprecate groupby.agg() with a dictionary when renaming).
# source: https://pandas.pydata.org/pandas-docs/stable/whatsnew/v0.25.0.html
```

#### 6. 获取每个小区的房源数量，并按从大到小的顺序排序

```python
df['xiaoqu_name'].value_counts().sort_values(ascending=False)
```

#### 7. 请统计出上海每个小区每平方米的均价

```python
df.groupby('xiaoqu_name')['up_price'].agg(pm_price ='mean').sort_values(by = 'pm_price', ascending = False)

# 或者：
# df.groupby('xiaoqu_name')['up_price'].mean().sort_values(ascending = False)
```

#### 8. 请统计出上海二手房总价排在前 5 位的房源，看看超级豪宅有多壕？

```python
df.sort_values(by='price',ascending = False).head(5)
```

#### 9. 请计算出不同年代的房子数量（10年一个区间）（任务提示：先将 buildyear 列转为时间序列，再利用 groupby 函数统计出每年的房源数量，最后用 resample 函数进行汇总。）

```python
type(df.buildyear[0])                                          #可以先查看数据类型
df.buildyear = pd.to_datetime(df.buildyear,format='%Y')        #再用pd.to_datetime转换成datetime类型
```

```python
# Method 1: 对于DataFrame，可直接用resample()函数的内置参数on选择要使用的列而不是索引进行重新采样。列必须与日期时间相似的数据。
# 这里的时间段选取10A是year end,如果是year start用10AS
df.resample('10A', on='buildyear').title.count()               
```

```python
# Method 2：把已经转换成datetime类型的列重新设置为索引，再用resample函数根据时间段汇总
df.set_index('buildyear').title.resample('10A').count()
```

```python
# Method 3：先用groupby根据年份分组并且用agg()统计每年的房源总数,再用resample函数进行汇总（这里的汇总是每十年相加）
df.groupby('buildyear')['title'].agg(amount = 'count').resample('10A').sum()
```
